```python
import pandas as pd
import plotly.express as px
```

# 대전 구별 인구데이터 전처리

```python
df = pd.read_csv('/content/drive/MyDrive/대전_인구데이터.csv', encoding='euc-kr')
df
```

```python
df['행정구역'] = df['행정구역'].str.replace(r'\s*\(.*\)\s*', '', regex=True)
df
```

```python
senior_df = df[['행정구역', '2023년_계_연령구간인구수']]
df['2023년_계_연령구간인구수'] = df['2023년_계_연령구간인구수'].astype(str).str.replace(',', '').astype(int)
senior_df = senior_df[df['행정구역'] != '대전광역시']
senior_df
```

```python
# '2024년07월_전체' 값을 기준으로 행정구역을 내림차순으로 정렬
category_order = senior_df.sort_values(by='2023년_계_연령구간인구수', ascending=False)['행정구역'].tolist()

# 가로 그래프
fig = px.histogram(senior_df, x='행정구역', y='2023년_계_연령구간인구수', text_auto=True,
                   category_orders={'행정구역': category_order})

fig.show()
```

```python
fig = px.pie(senior_df, names='행정구역', values='2023년_계_연령구간인구수')

fig.show()
```

#대전광역시 서구 데이터 전처리

```python
detail_df = pd.read_csv('/content/drive/MyDrive/대전_세부데이터.csv', encoding='euc-kr')
detail_df['행정구역'] = detail_df['행정구역'].str.replace(r'\s*\(.*\)\s*', '', regex=True)

detail_df
```

```python
# 필요한 컬럼을 숫자형으로 변환
columns_to_convert = [
    '2023년_계_총인구수',
    '2023년_계_연령구간인구수',
    '2023년_계_65~69세',
    '2023년_계_70~74세',
    '2023년_계_75~79세',
    '2023년_계_80~84세',
    '2023년_계_85~89세',
    '2023년_계_90~94세',
    '2023년_계_95~99세'
]

for column in columns_to_convert:
    detail_df[column] = detail_df[column].astype(str).str.replace(',', '').astype(int)

detail_df.head(5)
```

```python
detail_df['65세 이상 인구 수'] = (
    detail_df['2023년_계_65~69세'] +
    detail_df['2023년_계_70~74세'] +
    detail_df['2023년_계_75~79세'] +
    detail_df['2023년_계_80~84세'] +
    detail_df['2023년_계_85~89세'] +
    detail_df['2023년_계_90~94세'] +
    detail_df['2023년_계_95~99세']
)
detail_df.head(5)
```

```python
detail_senior_df = detail_df[['행정구역', '65세 이상 인구 수']]
detail_senior_df_sorted = detail_senior_df.sort_values(by='65세 이상 인구 수', ascending=False)

# 특정 행정구역을 제외한 데이터프레임 생성
excluded_regions = ['대전광역시', '대전광역시 서구', '대전광역시 중구', '대전광역시 동구', '대전광역시 유성구', '대전광역시 대덕구']
detail_senior_df_filtered = detail_senior_df_sorted[~detail_senior_df_sorted['행정구역'].isin(excluded_regions)]

# 결과 확인
detail_senior_df_filtered.head(20)
```

```python
# '대전광역시 동구'를 포함한 행정구역 필터링
filtered_df = detail_senior_df_filtered[detail_senior_df_filtered["행정구역"].str.contains("대전광역시 서구")]

#import ace_tools as tools; tools.display_dataframe_to_user(name="Filtered DataFrame", dataframe=filtered_df)

# 결과 확인
filtered_df.head(3)
```
